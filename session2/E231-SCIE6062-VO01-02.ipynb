```python
from Bio.Seq import Seq
from Bio.SeqUtils import molecular_weight
from Bio.SeqUtils import MeltingTemp as mt
from Bio.SeqUtils import GC
import matplotlib.pyplot as plt


dna_a = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC")
dna_b = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")

# Molecular Weight
print(molecular_weight(dna_a)) 
print(molecular_weight(dna_b)) 


# Melting Point
mt_a = mt.Tm_Wallace(dna_a)
mt_b = mt.Tm_Wallace(dna_b)

plt.bar(["Sequence A", "Sequence B" ], [mt_a,mt_b])
plt.xlabel("")
plt.ylabel("Melting Point")
plt.show()
# GC Content and Sequence A
importGCA = GC(dna_a)
print(importGCA)
importGCB = GC(dna_b)
print(importGCB)


ATA = (dna_a.count('A') + dna_a.count('T')) / len(dna_a) * 100
ATB = (dna_b.count('A') + dna_b.count('T')) / len(dna_b) * 100

plt.bar(["AT A", "AT B" ], [ATA,ATB])
plt.bar(["GC A", "GC B" ], [importGCA,importGCB])
plt.show()


print(ATA)
print(ATB)


```

Cara Kokonya

```python
GCSeqA = [x for x in dna_a if x in 'GC' ]
GCSeqB = [x for x in dna_b if x in 'GC' ]

RatioGCA = len(GCSeqA) / len(dna_a) * 100
RatioGCB = len(GCSeqB) / len(dna_b) * 100

print(RatioGCA)
print(RatioGCB)
```

```python
ATSeqA = [x for x in dna_a if x in 'AT' ]
ATSeqB = [x for x in dna_b if x in 'AT' ]

ratioATA = len(ATSeqA) / len(dna_a) * 100
ratioATB = len(ATSeqB) / len(dna_b) * 100

print(ratioATA, ratioATB)
```

```python
GC = [RatioGCA,RatioGCB]
label =  ["A", "B"]
plt.bar(label,GC)
plt.show()

AT = [ratioATA,ratioATB]
plt.bar(label,AT)
plt.show()
```
