```python
from Bio.Seq import Seq

dna_seq = Seq("ACTAGCTGACTAGCGTAC")

print(dna_seq)
```

Translasi dan Transkripsi

DNA -> 2 Helix

DNA -> mRNA

Translasi -> Translate si mRNA

```python
dir(dna_seq)

print(dna_seq.complement())
print(dna_seq.transcribe())

# bintang tanda stop kodon
print(dna_seq.translate())
# bisa juga transcribe trus translate
print(dna_seq.transcribe().translate())

# dibalik aja reverse complement
print(dna_seq.reverse_complement())
```

```python
from Bio.PDB.PDBParser import PDBParser

parser = PDBParser()

structure = parser.get_structure("6LU7", "6lu7/6lu7.pdb")

model = structure[0]

for chain in model:
    print(chain)
    
    for residue in chain:
        for atom in residue:
            print(atom)
```
